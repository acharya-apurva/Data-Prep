## Dataset Description

This data was extracted from the 1994 Census bureau database by Ronny Kohavi and Barry Becker (Data Mining and Visualization, Silicon Graphics). It contains information about individual factors that might affect personal income. The prediction task is to determine whether a person makes over 50K a year. The dataset contains 15 variables and 48842 rows.  


## Variable Descriptions
 
-**age:** individual's age (continuous) 
* This variable is kept in the data-set as income is generally known to rise as age rises.

-**workclass:** individual's work category (Private, Self-emp-not-inc, Self-emp-inc, Federal-gov, Local-gov, -State-gov, Without-pay, Never-worked) 
* This variable is kept in the dataset as workclass can be a good predictor of a person's income. For example, private jobs are known to pay more than government jobs.

-**fnlwgt:** the sampling weight of corresponding observations (continuous) 
* The weight of a person could also affect their income. Some studies have shown that men with larger weight earn more until the point of obesity after which their weight drops. That is why this varable is kept in the dataset.

-**education:** individual's highest education degree (Bachelors, Some-college, 11th, HS-grad, Prof-school, Assoc-acdm, Assoc-voc, 9th, 7th-8th, 12th, Masters, 1st-4th, 10th, Doctorate, 5th-6th, -Preschool). 
* This variable is kept in the dataset as individual's highet education degree is known to greatly influence their income with the income increasing as people attain higher degrees.

-**education-num:** continuous. (Enough information is not provided about the attribute. The attribute is kept for now in order to understand what it is.)

-**marital-status:** individual's marital status (Married-civ-spouse, Divorced, Never-married, Separated, Widowed, Married-spouse-absent, Married-AF-spouse)
* The variable is kept in the dataset as a individual's maritial status seems like a category that might potentially lower or increase someone's income.

-**occupation:** individual's occupation (Tech-support, Craft-repair, Other-service, Sales, Exec-managerial, Prof-specialty, Handlers-cleaners, Machine-op-inspct, Adm-clerical, Farming-fishing, Transport-moving, Priv-house-serv, Protective-serv, Armed-Forces)
* The variable is kept in the dataset as an individual's occupation greatly influences their income.

-**relationship:** individual's relation in his/her family (Wife, Own-child, Husband, Not-in-family, Other-relative, Unmarried) 
* The variable is kept in the dataset as a individual's relationship seems like a category that might potentially lower or increase someone's income.

-**race:** individual's race (White, Asian-Pac-Islander, Amer-Indian-Eskimo, Other, Black)
* The variable is kept in the dataset as a individual's race is known affect their income.

-**sex:** individual's sex (Female, Male)
* The variable is kept in the dataset as a individual's sex is known affect their income.

-**capital-gain:** individual's profits due to investments (continuous)
* The variable is kept in the dataset as a individual's capital gain might be good factor in predicting income as it shows the amount of extra money individual is making.

-**capital-loss:** individual's losses due to investments (continuous)
* The variable is kept in the dataset as a individual's capital loss might be good factor in predicting income as it usually shows the amount of money individuals can afford to lose.

-**hours-per-week:** individual's working hour per week (continuous) 
* The variable is kept in the dataset as the number of work hours per week of a person definitely affects their income.

-**native-country:** individual's native country (United-States, Cambodia, England, Puerto-Rico, Canada, Germany, Outlying-US(Guam-USVI-etc), India, Japan, Greece, South, China, Cuba, Iran, Honduras, Philippines, Italy, Poland, Jamaica, Vietnam, Mexico, Portugal, Ireland, France, Dominican-Republic, Laos, Ecuador, Taiwan, Haiti, Columbia, Hungary, Guatemala, Nicaragua, Scotland, Thailand, Yugoslavia, El-Salvador, Trinadad&Tobago, Peru, Hong, Holand-Netherlands)
* The variable is kept in the dataset as people are known to have different average incomes depending on the country they live.

-**income:** incidvidual's income (>50K,<50K,=>50K,<=50K) This is the value we want to predict and thus it will be dropped.

## Adding outliers to the data

I am going to add outliers randomly to the age and educational-num attributes as any outlier in these categories would be very obvious.

```python
import random #to generate random numbers
df3 = df2.copy() 
for row in range(len(df3.axes[0])):# for every row the data set
    
    if (random.random() < 0.03): #since we only want 3% of our data to be outliers
        #to add a little variation to our outliers I am generating a random numbers
        age_random = random.randint(50,100)
        #decrease the mean of the age column by 3 times the standard deviation minus the 
        #random variable and store it to df3
        df3.loc[row,'age']= df['age'].mean() - 3*df['age'].std()- age_random
        # it will be easy to identify outliers in age as age cannot be less than 0
        
        #same thing as above but with the educational num variable
    if (random.random() < 0.03):
        education_random= random.randint(0,5)  
        df3.loc[row,'educational-num'] = df['educational-num'].mean() + 3*df['educational-num'].std() + education_random #it
        #will be very easy to see outliers as numbers above 16 = 'doctorate' don't mean anything in the context of educational
        #educational num
```

### Min-Max scaling

*Iteration 2 Part 2*

```python
dff3 = dff2.copy()
import matplotlib.pyplot as plt

# Starting with Data Normalization

from sklearn.preprocessing import MinMaxScaler

# This function plots histograms with density curves before and after data scaling. It takes the original dataframe, columns,
# dataframe after trasformation and the figure title as parameters.
def plot_histograms_density(df, columns, df_transformed, fig_title):
    fig, axs = plt.subplots(len(columns),2,figsize=(16,18),constrained_layout=True)
    fig.suptitle(fig_title, fontsize=16)
    axs = axs.flatten()
    i = 0 
    for c in columns:
        dff3[c].hist(ax=axs[i], density=True) # normalizes the density
        dff3[c].plot.density(ax=axs[i], title=c)
        df_transformed[c].hist(ax=axs[i+1], density=True, stacked=True) # normalizes the density
        df_transformed[c].plot.density(ax=axs[i+1], title=str(c+'_transformed'))
        i+=2

# all the columns that need to be scaled and plotted
val_cols_names = ['age','workclass','fnlwgt','educational-num','marital-status','occupation','relationship','race','gender','capital-gain','capital-loss','hours-per-week','native-country','income']

# transform the data 
scaler = MinMaxScaler()
min_max_tdf = scaler.fit_transform(dff3)
min_max_tdf = pd.DataFrame(min_max_tdf, index=dff3.index, columns=dff3.columns)

#then plot it
plot_histograms_density(dff3, val_cols_names, min_max_tdf, 'MinMax Scaler')

# Iteration 2 Part 2
```

### Baseline Strategy

Since most of the attributes in the data are skewed in nature, we are going to use median values of the categories in order to replace the missing values. This is going to be used as the baseline strategy for us to improve upon.

For compaing performance between the different Data scaling technique + Regressor, we are going to use RMSE. It is perfect in assessing how well a model fits a dataset as it is measured in the same units as the response variable.

*Iteration 2 Part 3*

```python
from sklearn.preprocessing import Normalizer
from sklearn.preprocessing import QuantileTransformer
from sklearn.preprocessing import RobustScaler
from sklearn.preprocessing import MaxAbsScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import FunctionTransformer
from sklearn import metrics


import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsRegressor

from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedKFold
from sklearn.model_selection import cross_val_score

import matplotlib.pyplot as plt

# We are going to use the dataframe that has not been scaled yet and test how accurate median would be to replace the missing 
# values.
workclass_median = dff4['workclass'].median()
occupation_median = dff4['occupation'].median()
native_median = dff4['native-country'].median()

rmses = []

#The function takes takes np arrays as parameters and returns a rmse value using the difference between the two arrays.
def rmse(actual, pred): 
    actual, pred = np.array(actual), np.array(pred)
    return np.sqrt(np.square(np.subtract(actual,pred)).mean())

#Since we have three columns where the values are missing.
actual1 = dff4['workclass']
pred1 = [workclass_median] * dff4['workclass'].size

actual2 = dff4['occupation']
pred2 = [occupation_median] * dff4['occupation'].size

actual3 = dff4['native-country']
pred3 = [native_median] * dff4['native-country'].size

rmses.append(rmse(actual1,pred1))
rmses.append(rmse(actual2,pred2))
rmses.append(rmse(actual3,pred3))

#Show all the rmses
rmses

# Iteration 2 Part 3
```

### Linear Regression

Now we are going to use linear regression.

*Iteration 2 Part 3*

```python
#All the scalers that we are going to use
scaler_names = ['MinMax', 'MaxAbs', 'Robust',  'Quantile','Z-score', 'Log', 'EqualWidth']
arr_scalers = [MinMaxScaler(),MaxAbsScaler(), RobustScaler(),   QuantileTransformer(n_quantiles=10, random_state=0, output_distribution='normal'),Normalizer(), FunctionTransformer(np.log1p), KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')]

#Results for the 3 categories that contain missing values 
results1 = {}
results2 = {}
results3 = {}

model = LinearRegression()

#for all the scalers mentianed above
for i in range (len(arr_scalers)):
    # X is all the other variables except the ones containing the missing value
    X = dff4.drop(['workclass','occupation','native-country'], axis=1).to_numpy()
    X_t = arr_scalers[i].fit_transform(X)
    # y1 is workclass, y2 is occupation, and y3 is native country
    y1 = dff4['workclass'].to_numpy()
    y2 = dff4['occupation'].to_numpy()
    y3 = dff4['native-country'].to_numpy()
    # Let's estimate the values for several rounds of crossvalidation
    cv = RepeatedKFold(n_splits=10, n_repeats=5, random_state=1)
    
    scores1 = cross_val_score(model, X_t, y1, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
    scores2 = cross_val_score(model, X_t, y2, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
    scores3 = cross_val_score(model, X_t, y3, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
    
    results1[scaler_names[i]] = [abs(elem) for elem in scores1]
    results2[scaler_names[i]] = [abs(elem) for elem in scores2]
    results3[scaler_names[i]] = [abs(elem) for elem in scores3]


cv = RepeatedKFold(n_splits=10, n_repeats=5, random_state=1)
scores1 = cross_val_score(model, X, y1, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
scores2 = cross_val_score(model, X, y2, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)
scores3 = cross_val_score(model, X, y3, scoring='neg_root_mean_squared_error', cv=cv, n_jobs=-1)

results1['Raw data'] = [abs(elem) for elem in scores1]
results2['Raw data'] = [abs(elem) for elem in scores2]
results3['Raw data'] = [abs(elem) for elem in scores3]

df_res = pd.DataFrame(results1)
boxplot = df_res.boxplot() 
boxplot.set_ylabel('MSE')
plt.show()
#First we are showing the results for the attribute: workclass

#Iteration 2 Part 3
```
